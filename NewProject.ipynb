```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['BMI']=df['BMI'].fillna(df['BMI'].median())
```

```python
df['health_insurance_price']=df['health_insurance_price'].fillna(df['health_insurance_price'].median())
```

```python
df.isnull().sum()
```

```python
df['BMI'].describe()
```

```python
q1=df['BMI'].describe()['25%']
q3=df['BMI'].describe()['75%']
iqr=q3-q1
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
```

```python
df[(df['BMI']>upper_limit) |( df['BMI']<lower_limit)] # outliers
```

```python
df[(df['BMI']<upper_limit) & ( df['BMI']>lower_limit)]
```

```python
df['BMI'] = df['BMI'].clip(lower_limit,upper_limit)
```

```python
px.box(df,x='BMI',width=1400,height=300)
```

```python
df.isnull().sum()
```

```python
px.box(df,x='health_insurance_price',width=1400,height=300)
```

```python
q1=df['health_insurance_price'].describe()['25%']
q3=df['health_insurance_price'].describe()['75%']
iqr=q3-q1
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
```

```python
df[(df['health_insurance_price']<lower_limit) | (df['health_insurance_price']>upper_limit)]
```

```python
df[(df['health_insurance_price']>lower_limit) & (df['health_insurance_price']<upper_limit)]
```

```python
df['health_insurance_price']=df['health_insurance_price'].clip(lower_limit,upper_limit)
```

```python
px.box(df,x='health_insurance_price',width=1400,height=300)
```

```python
df['health_insurance_price']=df['health_insurance_price'].fillna(df['health_insurance_price'].mean())
```
